```python
# ============================================
# âš™ï¸ CELL C â€” Dataset Copy + Class Map Save
# ============================================

import os, json, shutil, tensorflow as tf
from pathlib import Path

# Define correct paths
DRIVE_DATASET = Path("/content/drive/MyDrive/SmartCropAI_Project/Dataset/PlantVillage")
LOCAL_DATASET = Path("/content/dataset/PlantVillage")
PROJECT_PATH = Path("/content/drive/MyDrive/SmartCropAI_Project")

# Copy dataset from Drive to Colab for faster training
if not LOCAL_DATASET.exists():
    print("ğŸ“‚ Copying dataset from Drive to Colab (takes 1â€“2 mins)...")
    os.makedirs("/content/dataset", exist_ok=True)
    shutil.copytree(DRIVE_DATASET, LOCAL_DATASET)
else:
    print("âœ… Dataset already exists locally.")

# Verify dataset folders
classes = sorted(os.listdir(LOCAL_DATASET))
print(f"âœ… Found {len(classes)} classes:")
print(classes)

# Save class map (index â†” label) to Drive
class_map = {cls_name: idx for idx, cls_name in enumerate(classes)}
with open(PROJECT_PATH / "class_map.json", "w") as f:
    json.dump(class_map, f, indent=2)
print(f"\nâœ… Class map saved to: {PROJECT_PATH}/class_map.json")

# TensorFlow check
print("\nTensorFlow:", tf.__version__)
print("GPU available:", tf.config.list_physical_devices('GPU'))
```

```python
# ============================================
# âš™ï¸ CELL D â€” DATA GENERATORS (TRAIN & VALIDATION)
# ============================================

from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = (128, 128)
BATCH_SIZE = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,   # 80% train, 20% validation
    rotation_range=20,
    zoom_range=0.2,
    horizontal_flip=True
)

train_gen = train_datagen.flow_from_directory(
    "/content/dataset/PlantVillage",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="training"
)

val_gen = train_datagen.flow_from_directory(
    "/content/dataset/PlantVillage",
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode="categorical",
    subset="validation"
)

print(f"\nâœ… Found {train_gen.samples} training images belonging to {train_gen.num_classes} classes.")
print(f"âœ… Found {val_gen.samples} validation images belonging to {val_gen.num_classes} classes.")
```

```python
# ============================================
# ğŸ§  CELL E â€” MODEL ARCHITECTURE & COMPILE
# ============================================

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Conv2D(128, (3,3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D(2,2),

    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),

    Dense(128, activation='relu'),
    Dropout(0.3),

    Dense(15, activation='softmax')  # 15 output classes
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
# ============================================
# ğŸš€ CELL F â€” MODEL TRAINING
# ============================================

from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import os

# Callbacks to prevent overfitting and auto-save best model
callbacks = [
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1),
    ModelCheckpoint(
        filepath="/content/drive/MyDrive/SmartCropAI_Project/Models/SmartCropAI_BestModel.h5",
        save_best_only=True,
        monitor='val_accuracy',
        mode='max',
        verbose=1
    )
]

# Train the model
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=20,
    callbacks=callbacks
)

# Save final trained model
final_model_path = "/content/drive/MyDrive/SmartCropAI_Project/Models/SmartCropAI_Final.h5"
os.makedirs(os.path.dirname(final_model_path), exist_ok=True)
model.save(final_model_path)

print(f"âœ… Training complete. Final model saved at: {final_model_path}")
```

```python
# Step B â€” Build MobileNetV2 model (correct preprocessing)
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import MobileNetV2, mobilenet_v2
import os, json

IMG_SHAPE = (128, 128, 3)
NUM_CLASSES = 15

# Create model folder if not already present
MODEL_SAVE_DIR = "/content/drive/MyDrive/SmartCropAI_Project/Models"
os.makedirs(MODEL_SAVE_DIR, exist_ok=True)

BEST_MODEL_PATH = os.path.join(MODEL_SAVE_DIR, "SmartCropAI_MobileNet_Best.h5")
FINAL_MODEL_PATH = os.path.join(MODEL_SAVE_DIR, "SmartCropAI_MobileNet_Final.keras")

# Load pre-trained MobileNetV2 as base
base_model = MobileNetV2(input_shape=IMG_SHAPE, include_top=False, weights='imagenet')
base_model.trainable = False  # freeze all layers initially

# Build final model
model = models.Sequential([
    layers.Input(shape=IMG_SHAPE),
    layers.Lambda(lambda img: mobilenet_v2.preprocess_input(img)),  # proper MobileNet preprocessing
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.3),
    layers.Dense(NUM_CLASSES, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=5e-4),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

```python
# Cell C â€” Data generators + model training (phase 1)
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping

BATCH_SIZE = 32
IMG_SIZE = (128, 128)

# Augmentation + rescale
train_datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # 80-20 split
)

train_gen = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    subset='training',
    class_mode='categorical'
)

val_gen = train_datagen.flow_from_directory(
    DATA_DIR,
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    subset='validation',
    class_mode='categorical'
)

# Callbacks for smart training
callbacks = [
    ModelCheckpoint(BEST_MODEL_PATH, save_best_only=True, monitor='val_accuracy', mode='max', verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1),
    EarlyStopping(monitor='val_accuracy', patience=5, restore_best_weights=True, verbose=1)
]

# Train only the top classifier layer first
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10,             # keep small first phase
    callbacks=callbacks
)

# Save final model
model.save(FINAL_MODEL_PATH)
print("âœ… Training complete. Final model saved at:", FINAL_MODEL_PATH)
```

```python
# Step D â€” Fine-tuning the last 30 layers of MobileNetV2
import tensorflow as tf

# Unfreeze the last 30 layers of base model
base_model.trainable = True
for layer in base_model.layers[:-30]:
    layer.trainable = False

print("âœ… Last 30 layers unfrozen for fine-tuning.")

# Recompile with very low LR
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Continue training
fine_tune_history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=5,       # small number, since base model is now partially trainable
    callbacks=callbacks
)

# Save updated fine-tuned model
FINE_TUNED_PATH = os.path.join(MODEL_SAVE_DIR, "SmartCropAI_MobileNet_FineTuned.keras")
model.save(FINE_TUNED_PATH)
print("ğŸ¯ Fine-tuning complete. Model saved at:", FINE_TUNED_PATH)
```
